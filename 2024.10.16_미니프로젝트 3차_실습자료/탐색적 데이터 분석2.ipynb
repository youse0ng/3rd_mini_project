```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import joblib
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
data_train = pd.read_csv('../2024.10.16_미니프로젝트 3차_실습자료\data01_train.csv')
data_test = pd.read_csv('../2024.10.16_미니프로젝트 3차_실습자료\data01_test.csv')
data_is_dy_train = joblib.load('../2024.10.16_미니프로젝트 3차_실습자료\Is_Dynamic_DF.pkl')
feature = pd.read_csv('../2024.10.16_미니프로젝트 3차_실습자료/features.csv')
data_test.drop(columns='subject',inplace=True)
data_train.drop(columns='subject',inplace=True)
```

```python
data = pd.read_csv('../2024.10.16_미니프로젝트 3차_실습자료\data01_train.csv')
data = data.drop(columns='subject')

LAYING = data.loc[data['Activity']=='LAYING']
SITTING = data.loc[data['Activity']=='SITTING']
WALKING = data.loc[data['Activity']=='WALKING']
WALKING_UPSTAIRS = data.loc[data['Activity']=='WALKING_UPSTAIRS']
WALKING_DOWNSTAIRS = data.loc[data['Activity']=='WALKING_DOWNSTAIRS']
STANDING = data.loc[data['Activity']=='STANDING']
IS_DYNAMIC = data_is_dy_train.loc[data_is_dy_train['Activity']==1]
IS_STATIC = data_is_dy_train.loc[data_is_dy_train['Activity']==0]
sensor_feature_importance_df = joblib.load('../2024.10.16_미니프로젝트 3차_실습자료\sensor_feature_importance_df.pkl')

LAYING.reset_index(inplace=True)
SITTING.reset_index(inplace=True)
WALKING.reset_index(inplace=True)
WALKING_UPSTAIRS.reset_index(inplace=True)
WALKING_DOWNSTAIRS.reset_index(inplace=True)
STANDING.reset_index(inplace=True)
```

```python
fig = px.bar(sensor_feature_importance_df,x='feature_importance',y='feature_sensor',
             color='feature_sensor',hover_data=['feature_importance'])
fig.update_layout(margin_l=50,
    margin_r=50,
    margin_b=50,
    margin_t=50,
    title_text='6가지 행동 분류 | 센서별 종합 피쳐 중요도',
    title_x=0.5,
    title_font_size = 25,
    title_font_color = "red",
    title_font_family = "Times",
    height=500,
    width=900
)
fig.update_xaxes(title_text='Importances')
fig.update_yaxes(title_text='Features')
fig.show()
```

```python
# 서브플롯 생성 (6행 1열)
fig = make_subplots(rows=6, cols=1,shared_xaxes=True,
                    subplot_titles=('WALKING', 'LAYING', 'SITTING', 
                                    'WALKING_DOWNSTAIRS', 'WALKING_UPSTAIRS', 'STANDING'))

# 각 서브플롯에 trace 추가
fig.add_trace(go.Scatter(y=WALKING['tGravityAcc-mean()-Y'], mode='lines', name='WALKING'), row=1, col=1)
fig.add_trace(go.Scatter(y=LAYING['tGravityAcc-mean()-Y'], mode='lines', name='LAYING'), row=2, col=1)
fig.add_trace(go.Scatter(y=SITTING['tGravityAcc-mean()-X'], mode='lines', name='SITTING'), row=3, col=1)
fig.add_trace(go.Scatter(y=WALKING_DOWNSTAIRS['tGravityAcc-mean()-X'], mode='lines', name='WALKING_DOWNSTAIRS'), row=4, col=1)
fig.add_trace(go.Scatter(y=WALKING_UPSTAIRS['tGravityAcc-mean()-X'], mode='lines', name='WALKING_UPSTAIRS'), row=5, col=1)
fig.add_trace(go.Scatter(y=STANDING['tGravityAcc-mean()-X'], mode='lines', name='STANDING'), row=6, col=1)

# 레이아웃 업데이트
fig.update_layout(title='tGravityAcc - X Axis',
                  xaxis_title='Time',
                  yaxis_title='GravityAcc',
                  legend_title='Activity',
                  height=800)  # 높이 조정
fig.show()
```
